```python
import pandas as pd

df = pd.read_csv('../data/airport-codes.csv')

df = df.rename(columns={'municipality': 'city'})

display(df[df['iata_code'] == "GVA"])
```

```python
minimal_df = df[['continent', 'iso_country', 'city', 'iata_code']]
```

```python
countries = pd.read_csv('../data/country-isocodes.csv')
```

```python
# Perform the join on the 'iso_country' column
merged_df = minimal_df.merge(countries[['iso_country', 'country']], on='iso_country', how='left')

display(merged_df)
```

```python
# An IATA airport code, also known as an IATA location identifier, 
# IATA station code, or simply a location identifier, is a three-letter code
# Only return airports with a iata_code for searching on the web
merged_df[merged_df['iata_code'].notnull()].to_csv('../data/airport-codes-minimal.csv', index=False)
```
